# Historical Trends in Physical Attributes

```python
#Define the function to analyze average trends
def average_trend(col):
    #Group the players by team and season
    #Find the average of the that particular column
    team_players = pd.DataFrame(df.groupby(['team_abbreviation','season'])[col].mean())
    #Rename according to the average column
    team_players = team_players.rename({'player_name': f'Average{col}'} ,axis=1)
    #Resets the old index and adds it as a column
    #Provide the overview of the data
    team_players.reset_index().head()
    
    #Creates a new figure
    plt.figure(figsize=(20,10))
    
    #Generate the lineplot
    sns.lineplot(x='season',y = col, data=team_players)
    
    #Set the title
    #Split it from the average column
    title = (' '.join(col.split('_'))).title()

    plt.xticks(rotation=-20)
    plt.ylabel(title, fontsize= 20)
    plt.title(f'Average {title} per Season', fontsize= 25)
 
    plt.show()
```

```python
#Height
average_trend('player_height')
```

Note: The average height has decreased over the years due to shift from Big Man game where taller players in positions like Center and Power forwards used to dominate the game to a more 'small ball' lineup with more focus on 3-point shooting

```python
#Age
average_trend('age')
```

Teams are more investing in younger players.

```python
#Weight
average_trend('player_weight')
```

Due more scientific advancement, better training and better nutrition, the average player weight has decreased over the years.

# Historical Top Performers

Let's find the historically top performing players by points, assists and rebounds in the NBA.

```python
#Top players by points
top_scorers = df.groupby('season')['pts'].max()
top_scorers = top_scorers.reset_index().merge(df[['season', 'player_name', 'pts']], on=['season', 'pts'], how='left')

for season in top_scorers['season'].unique():
    season_top_scorer = top_scorers[top_scorers['season'] == season]
    print(f"Season {season} Top Scorer: {season_top_scorer['player_name'].values[0]} - {season_top_scorer['pts'].values[0]} points")
```

```python
#Top players by assists
top_scorers = df.groupby('season')['ast'].max()
top_scorers = top_scorers.reset_index().merge(df[['season', 'player_name', 'ast']], on=['season', 'ast'], how='left')

for season in top_scorers['season'].unique():
    season_top_scorer = top_scorers[top_scorers['season'] == season]
    print(f"Season {season} Top Playmaker: {season_top_scorer['player_name'].values[0]} - {season_top_scorer['ast'].values[0]} assists")
```

```python
#Top players by rebounds
top_scorers = df.groupby('season')['reb'].max()
top_scorers = top_scorers.reset_index().merge(df[['season', 'player_name', 'reb']], on=['season', 'reb'], how='left')

for season in top_scorers['season'].unique():
    season_top_scorer = top_scorers[top_scorers['season'] == season]
    print(f"Season {season} Top Rebounder: {season_top_scorer['player_name'].values[0]} - {season_top_scorer['reb'].values[0]} rebounds")
```

Lets find the teams with most number of lottery picks. Lottery picks in the NBA are the draft picks where teams that didn't make the playoffs to determine the draft orders. Only the top 4 picks are decided by lottery.

```python
#Teams with most number of lottery picks
players_draft_year = df[df['draft_year'] == df['season'].apply(lambda x: x[:4])]

lottery_picks = players_draft_year[players_draft_year.draft_number.isin(['1','2','3','4'])]

lottery_picks.head()
```

```python
team_lottery_picks = lottery_picks['team_abbreviation'].value_counts()

plt.figure(figsize=(15,9))

sns.barplot(x = team_lottery_picks, y = team_lottery_picks.index)

plt.xlabel('Number of Lottery Picks (1996-2021)')
plt.xticks([i for i in range(1,10)])

plt.show()
```

# Top Colleges

Since college plays an important role in the draft, let's find the Top 10 colleges by total points, assists and rebounds.

```python
#Top colleges by points
#Group the colleges by total points
#Sort them in descending order
top_college_points = df.groupby('college')['pts'].sum().sort_values(ascending=False)
#Drop the none values
top_college_points = top_college_points.drop('None',errors= 'ignore')
#Take only the top 10 values
top_college_points = top_college_points.head(10)

#Plot these values
plt.figure(figsize=(12, 6))
plt.bar(top_college_points.index, top_college_points.values)
plt.xlabel('College')
plt.ylabel('Total Points')
plt.title('Top 10 Colleges by Total Points in the NBA')
plt.xticks(rotation=30, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
#Top colleges by rebounds
#Group the colleges by total rebounds
#Sort them in descending order
top_college_rebounds = df.groupby('college')['reb'].sum().sort_values(ascending=False)
#Drop the none values
top_college_rebounds = top_college_rebounds.drop('None',errors= 'ignore')
#Take only the top 10 values
top_college_rebounds = top_college_rebounds.head(10)

#Plot these values
plt.figure(figsize=(12, 6))
plt.bar(top_college_rebounds.index, top_college_rebounds.values)
plt.xlabel('College')
plt.ylabel('Total Rebounds')
plt.title('Top 10 Colleges by Total Rebounds in the NBA')
plt.xticks(rotation=30, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
#Top college by assists
#Group the colleges by total assists
#Sort them in descending order
top_college_assists = df.groupby('college')['ast'].sum().sort_values(ascending=False)
#Drop the none values
top_college_assists = top_college_assists.drop('None',errors= 'ignore')
#Take only the top 10 values
top_college_assists = top_college_assists.head(10)

#Plot these values
plt.figure(figsize=(12, 6))
plt.bar(top_college_points.index, top_college_points.values)
plt.xlabel('College')
plt.ylabel('Total Assists')
plt.title('Top 10 Colleges by Total Assists in the NBA')
plt.xticks(rotation=30, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```
